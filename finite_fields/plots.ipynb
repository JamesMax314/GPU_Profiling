# Profiling Plots for Finite Field Polynomeal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cpu_timings_us = np.loadtxt("cpu_timings_us.csv")
gpu_timings_us = np.loadtxt("gpu_timings_us.csv")
```

```python
ax = plt.figure().gca()
ax.set_yscale('log')

xs = np.linspace(0, 500, 500)
ax.plot(xs, cpu_timings_us/1e6, label="CPU")
ax.plot(xs, gpu_timings_us/1e6, label="GPU")

ax2 = ax.twinx()
ax2.set_yscale('log')

ax2.plot(xs, cpu_timings_us/gpu_timings_us, c="g", label="GPU Speed Up")

ax.set_xlabel(r"Polynomial degree")
ax.set_ylabel(r"$t\, [s]$")
ax2.set_ylabel(r"Fractional Speed Up")

ax.legend(loc='lower center')
ax2.legend(loc='lower right')

plt.savefig("finiteFields", bbox_inches="tight", dpi=300)

plt.show()
```
