# Profiling Plots for Finite Field Polynomeal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cpu_timings_us = np.loadtxt("cpu_timings_us.csv")
gpu_timings_us = np.loadtxt("gpu_timings_us.csv")
```

```python
ax = plt.figure().gca()
ax.set_yscale('log')

xs = np.linspace(0, 500, 500)
ax.plot(xs, cpu_timings_us/1e6)
ax.plot(xs, gpu_timings_us/1e6)

plt.xlabel(r"Polynomeal degree")
plt.ylabel(r"$t\, [s]$")
plt.savefig("finiteFields", bbox_inches="tight", dpi=300)

plt.show()
```
